```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time

def plotFunction(plt, min, max, step):
  x_values = np.arange(min, max+step, step)
  y_values = [f(x) for x in x_values]

  plt.plot(x_values, y_values, label=f'f(x) = {function}')
  plt.legend()

def f(x):
  return eval(function)

def goldenSectionSearch(a, b):
  limit = 0.00001
  time = 0

  while True:
    d = goldenRatio * (b - a)
    x1 = a + d
    x2 = b - d
    fx1 = f(x1)
    fx2 = f(x2)

    addData(time, d, x1, x2, a, b)
    plotPoints(plt, x1, x2)

    if abs(x2 - x1) < limit:
      return fx1
      break

    if fx1 > fx2:
      b = x1
    elif fx2 > fx1:
      a = x2

    time += 1

def addData (time, d, x1, x2, a, b):
  data['Iteração'].append(time)
  data['d'].append(d)
  data['x1'].append(x1)
  data['f(x1)'].append(f(x1))
  data['x2'].append(x2)
  data['f(x2)'].append(f(x2))
  data['a'].append(a)
  data['b'].append(b)

def plotPoints(plt, x1, x2):
  plt.scatter(x1, 0, c = 'blue')
  plt.scatter(x2, 0, c = 'red')


data = {
        'Iteração': [],
        'd': [],
        'x1': [],
        'f(x1)': [],
        'x2': [],
        'f(x2)': [],
        'a': [],
        'b': [],
}
```

```python

function = "x**2 - 6*x + 15"
a = 0
b = 10
goldenRatio = (5**(1/2) - 1) / (2)

minimun = goldenSectionSearch(a, b)

print(f'Mínimo encontrado entre [{a}, {b}] = {minimun}\n')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Gráfico da função f(x) = {function}')
plt.grid(True)

plotFunction(plt, a, b, 0.1)
plt.show()
df = pd.DataFrame(data)
display(df)
```
