# Método da Bisseção

  O código abaixo implementa o método da bisseção para minimizar funções unidimensionais.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import time

def plotFunction(min, max, step):
  plt.xlabel('x')
  plt.ylabel('f(x)')
  plt.title(f'Gráfico da função f(x) = {function}')
  plt.grid(True)

  x_values = np.arange(min, max+step, step)
  y_values = [f(x) for x in x_values]

  plt.plot(x_values, y_values, label=f'f(x) = {function}')
  plt.legend()
  plt.show()

def f(x):
  return eval(function)

def deriv_f(x):
  return eval(deriv_function)

def bissectionSearch(a, b):
  k = 0
  limit = 0.2
  actual_a, actual_b = a, b
  next_a, next_b = 0, 0

  while True:
    lmbda_k = (actual_a + actual_b)/2

    addData(k, lmbda_k, actual_a, actual_b)

    if deriv_f(lmbda_k) == 0:
      return f(lmbda_k)
    elif deriv_f(lmbda_k) > 0:
      next_a = actual_a
      next_b = lmbda_k
    else:
      next_a = lmbda_k
      next_b = actual_b

    if (1/2)**k <= limit/(b-a):
      return f(lmbda_k)
    else:
      actual_a = next_a
      actual_b = next_b
      k += 1

def addData (time, x1, a, b):
  data['Iteração'].append(time)
  data['λ'].append(x1)
  data['f(λ)'].append(f(x1))
  data['a'].append(a)
  data['b'].append(b)

data = {
        'Iteração': [],
        'λ': [],
        'f(λ)': [],
        'a': [],
        'b': [],
}
```
